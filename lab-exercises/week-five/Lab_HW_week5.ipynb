## Part 1 (Figures)
1. Create a plot showing the standard deviation (numpy.std) of the inflammation data for each day across all patients.

```python
import numpy as np
data = np.loadtxt(fname='../week-four/inflammation-01.csv', delimiter = ',')
```

```python
import matplotlib.pyplot as plt
# get standard deviation with np
std_inflammation = np.std(data, axis=0)
# plot std
std_plot = plt.plot(std_inflammation)
plt.xlabel('Time (days)')
plt.ylabel('Standard deviation')
plt.title('Inflammation')
plt.show()
```

2. Modify the program to display the three plots on top of one another instead of side by side.

```python
# create a figure object to hold multiple figures in one
fig = plt.figure(figsize=(4.0,10.0))
fig.suptitle('Inflammation score', va='bottom')

# make variables to store each subplot that we want
sbplt1 = fig.add_subplot(3, 1, 1)
    # nrows, ncols, index
sbplt2 = fig.add_subplot(3, 1, 2)
sbplt3 = fig.add_subplot(3, 1, 3)

# set y-axes

sbplt1.set_ylabel('average')
sbplt1.plot(np.mean(data, axis=0))

sbplt2.set_ylabel('max')
sbplt2.plot(np.max(data, axis=0))

sbplt3.set_ylabel('min')
sbplt3.plot(np.min(data, axis=0))
sbplt2.set_xlabel('time(days)')

# tight layout, provides adjustments to make sure it fits properly
fig.tight_layout()

plt.show()
```

## Part 2 (loops)

1. Explain what a for loop does.\
A for loop can run the same process multiple times over an infinite number of variables without having to rewrite the command for each variable. It can iterate over a list or specified range and allow you to run the same process on different data.

Python has a built-in function called range that generates a sequence of numbers. range can accept 1, 2, or 3 parameters.

If one parameter is given, range generates a sequence of that length, starting at zero and incrementing by 1. For example, range(3) produces the numbers 0, 1, 2.\
If two parameters are given, range starts at the first and ends just before the second, incrementing by one. For example, range(2, 5) produces 2, 3, 4.\
If range is given 3 parameters, it starts at the first one, ends just before the second one, and increments by the third one. For example, range(3, 10, 2) produces 3, 5, 7, 9.
