```python
import csv
import re
from collections import defaultdict
```

```python
mystring = 'Tadauchi, Miyanaga & Dawut, 2005'
mystring2 = '(Linnaeus, 1758)'
mystring3 = 'Meade-Waldo, 1914'
mystring4 = 'Michener, 1986'

test_strings = [mystring, mystring2, mystring3, mystring4]
```

```python
# auth_regex = re.compile(r'\(?[a-zA-Z]+-?[a-zA-Z]+')
    # doesn't recognize year
    # doesn't recognize if name has two '-'
    # a-z won't capture letters with accents
auth_regex = re.compile(r'\(?([\w,\&\-\s]*),\s(\d{4})\)?')
    # re.compile(author, whitespace, 4 digits)
    # \(? - outside of capture group, code will match it, but not capture it
    # creates match group 0 and match group 1
    
# write a function for regex match
# take each record and return list with two items
# item 1: capture group 1
# item 2: capture group 2

def match_author(record):
    author_match = re.match(auth_regex, record)
    authors = author_match.group(1)
    year = author_match.group(2)
    authors = re.split(',|\&', authors)
    return[authors, year]
```

```python
from collections import defaultdict

author_dict = defaultdict(int)
year_dict = defaultdict(int)

for record in test_strings:
    author_year = match_author(record)
    year_dict[author_year.group(1)] += 1
```
