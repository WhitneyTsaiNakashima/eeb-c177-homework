```python
import numpy as np
data = np.loadtxt(fname='../week-four/inflammation-01.csv', delimiter = ',')
```

```python
print(data)
```

```python
import matplotlib.pyplot as plt
```

```python
# heat map
image = plt.imshow(data)
plt.show()
```

```python
# transform data to take mean for each day
ave_inflammation = np.mean(data, axis=0)
# plot line/scatter plot
ave_plot = plt.plot(ave_inflammation)
plt.xlabel('Time (days)')
plt.ylabel('Inflammation score')
plt.title('Mean inflammation each day')
plt.show()
```

```python
# maximum inflammation score for each day
max_plot = plt.plot(np.max(data, axis=0))
plt.xlabel('Time (days)')
plt.ylabel('Inflammation score')
plt.title('Max inflammation each day')
plt.show()
```

```python
min_plot = plt.plot(np.min(data,axis=0))
plt.xlabel('Time (days)')
plt.ylabel('Inflammation score')
plt.title('Min inflammation each day')
plt.show()
```

```python
# create a figure object to hold multiple figures in one
fig = plt.figure(figsize=(10.0,3.0))
fig.suptitle('Inflammation score', va='bottom')

# make variables to store each subplot that we want
sbplt1 = fig.add_subplot(1, 3, 1)
    # nrows, ncols, index
sbplt2 = fig.add_subplot(1, 3, 2)
sbplt3 = fig.add_subplot(1, 3, 3)

# set y-axes

sbplt1.set_ylabel('average')
sbplt1.plot(np.mean(data, axis=0), '--')

sbplt2.set_ylabel('max')
sbplt2.plot(np.max(data, axis=0), 'o-')
sbplt2.set_xlabel('time(days)')

sbplt3.set_ylabel('min')
sbplt3.plot(np.min(data, axis=0), '*')

# tight layout, provides adjustments to make sure it fits properly
fig.tight_layout()

plt.show()
```

```python
# create a figure object to hold multiple figures in one
fig = plt.figure(figsize=(4.0,10.0))
fig.suptitle('Inflammation score', va='bottom')

# make variables to store each subplot that we want
sbplt1 = fig.add_subplot(3, 1, 1)
    # nrows, ncols, index
sbplt2 = fig.add_subplot(3, 1, 2)
sbplt3 = fig.add_subplot(3, 1, 3)

# set y-axes

sbplt1.set_ylabel('average')
sbplt1.plot(np.mean(data, axis=0))

sbplt2.set_ylabel('max')
sbplt2.plot(np.max(data, axis=0))

sbplt3.set_ylabel('min')
sbplt3.plot(np.min(data, axis=0))
sbplt2.set_xlabel('time(days)')

# tight layout, provides adjustments to make sure it fits properly
fig.tight_layout()

plt.show()
```

```python
length = 0
for vowel in 'aeiou' :
    length = length + 1

print('There are ', length, 'vowels.')

# this function is wrapped into a function len
print(len('aeiou'))
```

```python
newword = 'oxygen'
counter = 0

for char in newword:
    counter = counter + 1
    print(counter, ':', char)

# need counter to start at 0, because the first thing the loop does is add one to counter
```

```python
# python has an easier way to do this
for counter, char in enumerate(newword):
    counter += 1
    # shorthand way of writing counter = counter + 1
    # order also matters
    print(counter, ':', char)
```

# Lists

```python
odds = [1, 3, 5, 7]
print('odds are:', odds)
```

```python
print('first element:', odds[0])
```

```python
print('last element:', odds[3])
# could also use length function
print('last element:', odds[len(odds)-1])
```

```python
# another shorthand way to do this
print('last element:', odds[-1])
```

```python
for num in odds:
    print(num)
```

```python
for ii in range(0, len(odds)):
    print(odds[ii])
```
